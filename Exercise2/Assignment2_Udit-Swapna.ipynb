### Exercise 3: Data analysis

Here is sample code to pass your data from the pandas data frame to bokeh and render a graphic inside the notebook.

```python
filename = "baseball_data.csv"
df = pd.read_csv( filename, header=0 )
```

```python
# compute a histogram of the height variable
df_height = df['height'].value_counts().to_frame('count')
df_height.head()
```

```python
# import bokeh 
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# tell bokeh to show the figures in the notebook
from bokeh.io import output_notebook
output_notebook()
```

```python
# create a figure
p = figure( plot_height=300, title="Height distribution of baseball players" )

# render the data
#p.vbar( source=df_height, x='height', width=1.0, bottom=0, top='count',
#        line_color='white', fill_color='cornflowerblue')
p.vbar( source=df_height, x='index', width=1.0, bottom=0, top='count',
        line_color='white', fill_color='cornflowerblue')


# annotate the graphic
p.xaxis.axis_label = "height [inches]"
p.yaxis.axis_label = "count"

p.xgrid.visible = False

show(p)
```
