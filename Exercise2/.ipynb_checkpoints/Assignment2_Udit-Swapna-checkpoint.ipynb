### Exercise 3: Data analysis

Here is sample code to pass your data from the pandas data frame to bokeh and render a graphic inside the notebook.

```python
filename = "titanic3.csv"
df = pd.read_csv( filename, header=0 )
```

```python
# compute a histogram of the height variable
df_age_survival = df[['survived','age']]
df_age_survival.age = df_age_survival.age.round()
df_survivals = df_age_survival[df_age_survival.survived > 0]
df_dead = df_age_survival[df_age_survival.survived < 1]
df_age_survived=df_survivals['age'].value_counts().to_frame('count')
df_age_dead=df_dead['age'].value_counts().to_frame('count')
#df_survivals.head()
df_age_survived.head()
df_age_dead.head()
```

```python
# import bokeh 
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# tell bokeh to show the figures in the notebook
from bokeh.io import output_notebook
output_notebook()
```

```python
# create a figure
p = figure( plot_height=300, title="Distribution of passengers in titanic survived by age" )
p1 = figure( plot_height=300, title="Distribution of passengers in titanic dead by age" )

# render the data
#p.vbar( source=df_height, x='height', width=1.0, bottom=0, top='count',
#        line_color='white', fill_color='cornflowerblue')
p.vbar( source=df_age_survived, x='index', width=1.0, bottom=0, top='count',
        line_color='white', fill_color='cornflowerblue')
p1.vbar( source=df_age_dead, x='index', width=1.0, bottom=0, top='count',
        line_color='white', fill_color='red')


# annotate the graphic
p.xaxis.axis_label = "Age [years]"
p.yaxis.axis_label = "count"

p.xgrid.visible = False

show(p)

p1.xaxis.axis_label = "Age [years]"
p1.yaxis.axis_label = "count"

p1.xgrid.visible = False

show(p1)
```

```python
df_class_survival = df[['survived','pclass']]
#df_age_survival.age = df_age_survival.age.round()
df_survivals = df_class_survival[df_class_survival.survived > 0]
df_dead = df_class_survival[df_class_survival.survived < 1]
df_class_survived=df_survivals['pclass'].value_counts().to_frame('count')
df_class_dead=df_dead['pclass'].value_counts().to_frame('count')
#df_survivals.head()
df_class_survived.head()
df_class_dead.head()
```

# Summary
The Titanic Dataset gives an insight into the number of people that survived the disaster with 0 (Not Survived) and 1 (Survived). Out of the total passengers that were on board the ship, less percentage of people surived where as majority of them could not survive the disaster.

Q1 Distributions of passengers who survived based on age shows that children survived more as compared to people above 35 years of age. For people with age between 18 and 35 years, ratio of survival and death is almost same. We could say that preference was given to save children first (based on age).

Q2 Distributions of passengers who survived based on class shows that people in class 1 survived more as compared to people in class 3.
The people travelling in each class was not the same. For this purpose, we can continue the exploration by taking the mean of values according to the class so that we can get a proper distribution of the same.

Q3 Distribution of passengers with respect to survival based on sex shows that female passengers were given preference over male passengers.

Q4 Distribution of age of passengers can be seen in the graph.We can conclude that the passengers between the age of 18 to 35 years are more.

Q5 Distribution of class shows that the number of passengers in Class 3 were more as compared to Class 1 and Class 2.

Q6 Distributions of passengers who survived based on age and class shows that majority of people survived in class 1.
Most of the children in class 1 survived the disaster as there were less children in that class. Majority of people in class 1 between the age 18 and 35 years survived. Majority of people above 35 years died in class 1. 
In Class 2, all the children survived the disaster. Majority of people in class 1 between the age 18 and 35 years survived. Majority of people above 35 years died in class 1. 
Number of survivals were very less than the number of deaths.

In total, we can conclude that children were given priority in all the classes. Survival of passengers was dependent on age, class and sex. Age and sex were the impacting factors.
